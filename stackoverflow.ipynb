# Visualization

```python
def plot_line_chart(df, column, line=''):
    if line == '':
        line = df[column].value_counts().keys()[0:20]
    data = df[column].value_counts()[0:20]
    fig = px.line(x=line, y=data)
    fig.show()
def plot_bar_chart(df, column, line=''):
    if line == '':
        line = df[column].value_counts().keys()[0:20]
    data = df[column].value_counts()[0:20]
    fig = px.bar(x=line, y=data)
    fig.show()

def plot_pie_chart(df, column, line=''):
    if line == '':
        line = df[column].value_counts().keys()[0:20]
    data = df[column].value_counts()[0:20]
    fig = px.pie(names=line, values=data)
    fig.show()    
```

```python
# def shorten_categories(categories, cutoff):
#     categorical_map = {}
#     for i in range(len(categories)):
#         if categories.values[i] >= cutoff:
#             categorical_map[categories.index[i]] = categories.index[i]
#         else:
#             categorical_map[categories.index[i]] = 'Other'
#     return categorical_map
def shorten_categories(categories, cutoff):
    return {
        category: category if value >= cutoff else "Other"
        for category, value in zip(categories.index, categories.values)
    }
```

```python
country_map = shorten_categories(df.Country.value_counts(), 199)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
def clean_experience(x):
    if x == 'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)
```

```python
# Apply the function
df['EdLevel'] = df['EdLevel'].apply(clean_education)

# Check the unique values to ensure proper categorization
print(df['EdLevel'].unique())
# print(df['EdLevel'].value_counts())
```

# Model Training

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X =df.drop('Salary', axis=1)
# X = df.iloc[:, :3]
y = df['Salary']
```

```python
# Numeric pipeline with imputer and scaler
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scale', StandardScaler())
])
# Categorical pipeline
categorical_pipeline = Pipeline([('OneHot', OneHotEncoder(handle_unknown='ignore'))])

# Column transformer
transformer = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, ['YearsCodePro']),
        ('cat', categorical_pipeline, ['Country', 'EdLevel'])
    ]
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.10, random_state=1234)
```

```python
X_train 
```

```python
mlpipe_1 = Pipeline([('Transformer',transformer), ('lr', LinearRegression())])
```

```python
mlpipe_1.fit(X_train, y_train)
```

```python
y_hat = mlpipe_1.predict(X_test)
y_hat
```

```python
from sklearn.metrics import r2_score, mean_squared_error
error = np.sqrt(mean_squared_error(y_test, y_hat))
print('${:,.02f}'.format(error))
```

# Training a new model using Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

mlpipe_2 = Pipeline([('Transformer',transformer), ('dt', DecisionTreeRegressor())])
```

```python
mlpipe_2.fit(X_train, y_train)
```

```python
y_hat = mlpipe_2.predict(X_test)
y_hat
```

```python
from sklearn.metrics import r2_score, mean_squared_error
error = np.sqrt(mean_squared_error(y_test, y_hat))
print('${:,.02f}'.format(error))
```

# Training a new model using RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
mlpipe_3 = Pipeline([('Transformer', transformer), ('rf', RandomForestRegressor())])
mlpipe_3.fit(X_train, y_train)
y_hat = mlpipe_3.predict(X_test)
y_hat
```

```python
error = np.sqrt(mean_squared_error(y_test, y_hat))
print('${:,.02f}'.format(error))
```

# Training a new model using XGBRegressor

```python
from xgboost import XGBRegressor
mlpipe_4 = Pipeline([('Transformer', transformer), ('xgb', XGBRegressor())])
mlpipe_4.fit(X_train, y_train)
y_hat = mlpipe_4.predict(X_test)
y_hat
```

```python
error = np.sqrt(mean_squared_error(y_test, y_hat))
print('${:,.02f}'.format(error))
```
